# Model Optimization

Example notebook for exploring model optimization, using scikit-learn's GridSearchCV and RandomizedSearchCV, for hyperparameter tuning.

```python
import pandas as pd
import numpy as np
```

```python
from matplotlib import pyplot as plt
from sklearn.datasets import fetch_covtype
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
RANDOM_SEED = 7657

TEST_SIZE=0.80
```

#### Loading dataset

Forest Cover Types dataset

```python
cover = fetch_covtype()
```

```python
print(cover.DESCR)
```

```python
cover.target = cover.target-1
```

```python
set(cover.target)
```

```python
class_names = {
    0: "0: Lodgepole Pine",
    1: "1: Spruce/Fir",
    2: "2: Ponderosa Pine",
    3: "3: Cottonwood/Willow",
    4: "4: Aspen",
    5: "5: Douglas-fir",
    6: "6: Krummholz"
}
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    cover.data, cover.target, random_state=RANDOM_SEED, test_size=TEST_SIZE)
```

```python
print(f"Train set size: {X_train.shape[0]}")
print(f"Test set size:  {X_test.shape[0]}")
```

### Hyperparameter tuning with Grid and Random search

Details on [GridSearchCV](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html) and [RandomizedSearchCV](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.RandomizedSearchCV.html)

Details on [DecisionTreeClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html)

```python
tree = DecisionTreeClassifier(random_state=RANDOM_SEED)
```

```python
hyper_parameters = {
    'parameter_1': [..., ..., ...],
    'parameter_1': [..., ..., ...],
    }
```

```python
clf = RandomizedSearchCV(
    estimator = tree, 
    param_distributions = hyper_parameters,
    n_iter=20,
    scoring = 'accuracy',
    cv = 5, # Determines the cross-validation splitting strategy
    verbose=True,
    n_jobs=-1)
```

```python
# OR you can use GridSearchCV !
#
#clf = GridSearchCV(
#)
```

```python
clf.fit(X_train,y_train)
```

```python
# Estimator that was chosen by the search, i.e. estimator which gave highest score (or smallest loss if specified) on the left out data.

best_model = clf.best_estimator_
```

```python
# Accuracy on the best model

print("Accuracy on train set:  {:.1f}%".format(best_model.score(X_train, y_train)*100))
print("Accuracy on test set:   {:.1f}%".format(best_model.score(X_test, y_test)*100))
```
