# Visualization with Plotly

Plotly is a library that allows you to create interactive plots that you can use in dashboards or websites (you can save them as html files or static images).

Source:
based on https://github.com/vharivinay/python-eda-viz/tree/main

## Imports and Set-up

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

### Fake Data

```python
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})
```

```python
df2.head()
```

## Using iplot()

* scatter
* bar
* box
* ratio
* heatmap
* surface
* histogram
* bubble

## Scatter

```python
fig = px.scatter(df, x='A', y='B')
fig.show()
```

## Bar Plots

```python
fig = px.bar(df2, x='Category', y='Values')
fig.show()
```

```python
counts = df.count()
fig = px.bar(x=counts.index, y=counts.values, labels={'x': 'Columns', 'y': 'Count'})
fig.show()
```

## Boxplots

```python
fig = px.box(df)
fig.show()
```

## 3D Surface

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})

fig = go.Figure(data=[go.Surface(z=df3.values)])
fig.update_layout(scene=dict(zaxis_title='Value'))
fig.show()
```

## Histogram

```python
fig = px.histogram(df, x='A', nbins=20)
fig.show()
```

## Bubble plot / Scatter Matrix

Similar to sns.pairplot()

```python
fig = px.scatter(df, x='A', y='B', size=df['C'].abs())
fig.show()
```

## Scatter Matrix

Similar to sns.pairplot()

```python
fig = px.scatter_matrix(df, dimensions=['A', 'B', 'C'], color='D')
fig.show()
```
